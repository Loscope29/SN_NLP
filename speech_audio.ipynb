```python
import edge_tts
from IPython.display import Audio, display
```

```python
async def tts_notebook(text, voice="fr-FR-DeniseNeural"):
    """
    Génère l'audio et l'affiche directement dans le notebook.
    """
    output_file = "jola.mp3"
    
    # Configuration et génération
    communicate = edge_tts.Communicate(text, voice)
    await communicate.save(output_file)
    
    # Affichage du lecteur audio dans le notebook
    display(Audio(output_file, autoplay=True))

```

```python
resultat_rag = "D'après vos documents, le projet a été validé le 12 mars dernier."
```

```python
await tts_notebook(resultat_rag)
```
